## Visualization of 1D Data

Draw a histogram and a box and whisker plot of the printed data on paper

```python
import numpy as np
s = np.random.randint(5,25,35)
hist, bin_edges = np.histogram(s, bins=10)

print("Histogram:", hist)
print("Bin edges:", bin_edges)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your dataset
data = [
    20, 20, 16, 20, 16, 11, 18, 19, 24, 20, 15, 18, 9, 5, 12, 23, 16, 8, 15, 11,
    22, 6, 12, 12, 13, 14, 19, 17, 24, 22, 14, 22, 6, 15, 7
]

# Create a figure with two subplots, arranged vertically
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

# --- Plot 1: Histogram ---
ax1.hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
ax1.set_title('Histogram of the Dataset')
ax1.set_xlabel('Value')
ax1.set_ylabel('Frequency')
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# --- Plot 2: Box and Whisker Diagram ---
ax2.boxplot(data, vert=False, patch_artist=True,
            boxprops=dict(facecolor='lightgreen', color='black'),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'),
            medianprops=dict(color='red', linewidth=2))
ax2.set_title('Box and Whisker Diagram of the Dataset')
ax2.set_xlabel('Value')
ax2.set_yticks([]) # Hide the y-axis ticks for a cleaner look
ax2.grid(axis='x', linestyle='--', alpha=0.7)

# Adjust layout to prevent plots from overlapping
plt.tight_layout()

# Show the combined plot
plt.show()
```
